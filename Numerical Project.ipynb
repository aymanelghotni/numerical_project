```python
def cos(x):
    return math.cos(x)
def sin(x):
    return math.sin(x)
def e(x):
    return math.exp(x)
def tan(x):
    return math.tan(x)
def toFunction(text,x):
    return eval(text,{'__builtins__':{'cos': cos, "x":x,"sin":sin,"e":exp,"tan":tan}})
```

```python
def f(x):
    return x**2 + x + 1
```

```python
def bisectionMethod (xl, xu, Es):
    if (f(xl) * f(xu) >= 0):
        return false
    
    L0 = xu - xl
    n = log2(L0/(xl * Es))
    
    xr = 0
    
    for i in range (n):
        xr = (xl + xu)/2
        test = f(xl) * f(xr)
        
        if (test == 0):
            break
        elif (test > 0):
            xl = xr
        elif (test < 0):
            xu = xr
            
    return xr
```

```python
def falsePosition (xl, xu, Es):
    if (f(xl) * f(xu) >= 0):
        return false
    
    L0 = xu - xl
    n = log2(L0 / (xl * Es))
    
    xr = 0
    
    for i in range (n):
        xr = (xl * f(xu) - xu * f(xl)) / (f(xu) - f(xl))
        test = f(xr)
        
        if (test == 0):
            break
        elif (test > 0):
            xu = xr
        elif (test < 0):
            xl = xr
            
    return xr
```
